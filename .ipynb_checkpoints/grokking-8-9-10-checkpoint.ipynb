```python
import sys, numpy as np
from keras.datasets import mnist

np.random.seed(1)

(x_train, y_train), (x_test, y_test) = mnist.load_data()

#reshaping converts the 28x28 image matrix into a vector for easier calculation, 255 is RGB code of white.
images, labels = (x_train[0:1000].reshape(1000,28*28)/255, y_train[0:1000]) 
# print(f"{labels[0]} = {images[0]}") - THIS IS NUMBER 5

#create a vector 10 units wide, and set the corresponding value to 1 to indicate presence of a label prediction  
one_hot_labels = np.zeros((len(labels),10))
#print(one_hot_labels.shape)

for i,l in enumerate(labels): #enumerate function returns counter and value
    one_hot_labels[i][l] = 1
#    print(f"l={l} {one_hot_labels[i]}")
labels = one_hot_labels


test_images = x_test.reshape(len(x_test),28*28) / 255
test_labels = np.zeros((len(y_test),10))

#print(test_images.shape)

for i,l in enumerate(y_test):
    test_labels[i][l] = 1
    
relu = lambda x:(x>=0) * x # returns x if x > 0, return 0 otherwise
relu2deriv = lambda x: x>=0 # returns 1 for input > 0, return 0 otherwise
alpha, iterations, hidden_size, pixels_per_image, num_labels = (0.005, 30, 40, 784, 10)

weights_0_1 = 0.2*np.random.random((pixels_per_image,hidden_size)) - 0.1
weights_1_2 = 0.2*np.random.random((hidden_size,num_labels)) - 0.1

for j in range(iterations):
    error, correct_cnt = (0.0, 0)
    test_error, test_correct_cnt = (0.0, 0)
    
    for i in range(len(images)):
        layer_0 = images[i:i+1]
        layer_1 = relu(np.dot(layer_0,weights_0_1)) #Relu only returns values where weights > 0
        layer_2 = np.dot(layer_1,weights_1_2)

        error += np.sum((labels[i:i+1] - layer_2) ** 2)
        #print(f"np.argmax(layer_2)={np.argmax(layer_2)} np.argmax(labels[{i}:{1+1}]={np.argmax(labels[i:i+1])}")
        #print(np.argmax(layer_2) == np.argmax(labels[i:i+1]))
        #when the equality check returns True, this increments the correct_count
        correct_cnt += int(np.argmax(layer_2) == np.argmax(labels[i:i+1])) 
        #print(correct_cnt)
        
        layer_2_delta = (labels[i:i+1] - layer_2)
        layer_1_delta = layer_2_delta.dot(weights_1_2.T) * relu2deriv(layer_1)
        weights_1_2 += alpha * layer_1.T.dot(layer_2_delta)
        weights_0_1 += alpha * layer_0.T.dot(layer_1_delta)

#    sys.stdout.write("\r Training Data I:"+str(j)+ \
#                     " Train-Err:" + str(error/float(len(images)))[0:5] +\
#                     " Train-Acc:" + str(correct_cnt/float(len(images))))
    
#    if(j % 10 == 0 or j == iterations-1):
        
    for i in range(len(test_images)):

        layer_0 = test_images[i:i+1]
        layer_1 = relu(np.dot(layer_0,weights_0_1))
        layer_2 = np.dot(layer_1,weights_1_2)

        test_error += np.sum((test_labels[i:i+1] - layer_2) ** 2)
        test_correct_cnt += int(np.argmax(layer_2) == \
                                        np.argmax(test_labels[i:i+1]))
 
    sys.stdout.write("\r I:"+str(j)+ \
                     " Train-Err:" + str(error/float(len(images)))[0:5] +\
                     " Train-Acc:" + str(correct_cnt/float(len(images))) +\
                     " === Test-Err:" + str(test_error/float(len(test_images)))[0:5] +\
                     " Test-Acc:" + str(test_correct_cnt/float(len(test_images))))
   # print()
```

### Dropout Code

```python
import sys, numpy as np
from keras.datasets import mnist

np.random.seed(1)

(x_train, y_train), (x_test, y_test) = mnist.load_data()

#reshaping converts the 28x28 image matrix into a vector for easier calculation, 255 is RGB code of white.
images, labels = (x_train[0:1000].reshape(1000,28*28)/255, y_train[0:1000]) 
# print(f"{labels[0]} = {images[0]}") - THIS IS NUMBER 5

#create a vector 10 units wide, and set the corresponding value to 1 to indicate presence of a label prediction  
one_hot_labels = np.zeros((len(labels),10))
#print(one_hot_labels.shape)

for i,l in enumerate(labels): #enumerate function returns counter and value
    one_hot_labels[i][l] = 1
#    print(f"l={l} {one_hot_labels[i]}")
labels = one_hot_labels


test_images = x_test.reshape(len(x_test),28*28) / 255
test_labels = np.zeros((len(y_test),10))

#print(test_images.shape)

for i,l in enumerate(y_test):
    test_labels[i][l] = 1
    
relu = lambda x:(x>=0) * x # returns x if x > 0, return 0 otherwise
relu2deriv = lambda x: x>=0 # returns 1 for input > 0, return 0 otherwise
alpha, iterations, hidden_size, pixels_per_image, num_labels = (0.005, 300, 40, 784, 10)

weights_0_1 = 0.2*np.random.random((pixels_per_image,hidden_size)) - 0.1
weights_1_2 = 0.2*np.random.random((hidden_size,num_labels)) - 0.1

for j in range(iterations):
    error, correct_cnt = (0.0, 0)
    test_error, test_correct_cnt = (0.0, 0)
    
    for i in range(len(images)):
        layer_0 = images[i:i+1]
        layer_1 = relu(np.dot(layer_0,weights_0_1)) #Relu only returns values where weights > 0
        dropout_mask = np.random.randint(2,size=layer_1.shape)
        layer_1 = layer_1 * dropout_mask * 2 #multiply by 2 to keep sum of weighted values same, and not cut in half
        #print(dropout_mask)
        layer_2 = np.dot(layer_1,weights_1_2)

        error += np.sum((labels[i:i+1] - layer_2) ** 2)
        #print(f"np.argmax(layer_2)={np.argmax(layer_2)} np.argmax(labels[{i}:{1+1}]={np.argmax(labels[i:i+1])}")
        #print(np.argmax(layer_2) == np.argmax(labels[i:i+1]))
        #when the equality check returns True, this increments the correct_count
        correct_cnt += int(np.argmax(layer_2) == np.argmax(labels[i:i+1])) 
        #print(correct_cnt)
        
        layer_2_delta = (labels[i:i+1] - layer_2)
        layer_1_delta = layer_2_delta.dot(weights_1_2.T) * relu2deriv(layer_1)
        layer_1_delta = layer_1_delta * dropout_mask #remove 50% of error to stay consistent with 50% cut in input
        
        weights_1_2 += alpha * layer_1.T.dot(layer_2_delta)
        weights_0_1 += alpha * layer_0.T.dot(layer_1_delta)

#    sys.stdout.write("\r Training Data I:"+str(j)+ \
#                     " Train-Err:" + str(error/float(len(images)))[0:5] +\
#                     " Train-Acc:" + str(correct_cnt/float(len(images))))
    
#    if(j % 10 == 0 or j == iterations-1):
        
    for i in range(len(test_images)):

        layer_0 = test_images[i:i+1]
        layer_1 = relu(np.dot(layer_0,weights_0_1))
        layer_2 = np.dot(layer_1,weights_1_2)

        test_error += np.sum((test_labels[i:i+1] - layer_2) ** 2)
        test_correct_cnt += int(np.argmax(layer_2) == \
                                        np.argmax(test_labels[i:i+1]))
 
    sys.stdout.write("\r I:"+str(j)+ \
                     " Train-Err:" + str(error/float(len(images)))[0:5] +\
                     " Train-Acc:" + str(correct_cnt/float(len(images))) +\
                     " === Test-Err:" + str(test_error/float(len(test_images)))[0:5] +\
                     " Test-Acc:" + str(test_correct_cnt/float(len(test_images))))
   # print()
```

### Dropout Code + Batching of Gradient Descent algo

```python
import sys, numpy as np
from keras.datasets import mnist

np.random.seed(1)

relu = lambda x:(x>=0) * x # returns x if x > 0, return 0 otherwise
relu2deriv = lambda x: x>=0 # returns 1 for input > 0, return 0 otherwise

#Load the data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#reshaping converts the 28x28 image matrix into a vector for easier calculation, 255 is RGB code of white.
images, labels = (x_train[0:1000].reshape(1000,28*28)/255, y_train[0:1000]) 
# print(f"{labels[0]} = {images[0]}") - THIS IS NUMBER 5

#create a vector 10 units wide, and set the corresponding value to 1 to indicate presence of a label prediction  
one_hot_labels = np.zeros((len(labels),10))
#print(one_hot_labels.shape)

for i,l in enumerate(labels): #enumerate function returns counter and value
    one_hot_labels[i][l] = 1
#    print(f"l={l} {one_hot_labels[i]}")
labels = one_hot_labels


test_images = x_test.reshape(len(x_test),28*28) / 255
test_labels = np.zeros((len(y_test),10))

#print(test_images.shape)

for i,l in enumerate(y_test):
    test_labels[i][l] = 1
    
batch_size = 100
alpha, iterations = (.001, 300) 
hidden_size, pixels_per_image, num_labels = (100, 784, 10)

weights_0_1 = 0.2*np.random.random((pixels_per_image,hidden_size)) - 0.1
weights_1_2 = 0.2*np.random.random((hidden_size,num_labels)) - 0.1

for j in range(iterations):
    error, correct_cnt = (0.0, 0)
    test_error, test_correct_cnt = (0.0, 0)
    
    #TRAINING
    for i in range(int(len(images)) / batch_size):
        batch_start, batch_end = ((i * batch_size),((i+1)*batch_size))
        
        layer_0 = images[batch_start:batch_end]
        
        layer_1 = relu(np.dot(layer_0,weights_0_1)) #Relu only returns values where weights > 0
        dropout_mask = np.random.randint(2,size=layer_1.shape)
        layer_1 = layer_1 * dropout_mask * 2 #multiply by 2 to keep sum of weighted values same, and not cut in half
        #print(dropout_mask)
        layer_2 = np.dot(layer_1,weights_1_2)

        error += np.sum((labels[batch_start:batch_end] - layer_2) ** 2)
 
        #Lets update the weights for each batch size
        for k in range(batch_size):
            #when the equality check returns True, this increments the correct_count
            correct_cnt += int(np.argmax(layer_2[k:k+1]) == np.argmax(labels[batch_start+k:batch_start+k+1])) 
            #print(correct_cnt)

            layer_2_delta = (labels[batch_start:batch_end] - layer_2)/batch_size
            layer_1_delta = layer_2_delta.dot(weights_1_2.T) * relu2deriv(layer_1)
            layer_1_delta = layer_1_delta * dropout_mask #remove 50% of error to stay consistent with 50% cut in input

            weights_1_2 += alpha * layer_1.T.dot(layer_2_delta)
            weights_0_1 += alpha * layer_0.T.dot(layer_1_delta)

    #TESTING        
    for i in range(len(test_images)):

        layer_0 = test_images[i:i+1]
        layer_1 = relu(np.dot(layer_0,weights_0_1))
        layer_2 = np.dot(layer_1,weights_1_2)

        test_error += np.sum((test_labels[i:i+1] - layer_2) ** 2)
        test_correct_cnt += int(np.argmax(layer_2) == np.argmax(test_labels[i:i+1]))
 
    sys.stdout.write("\r I:"+str(j)+ \
                     " Train-Err:" + str(error/float(len(images)))[0:5] +\
                     " Train-Acc:" + str(correct_cnt/float(len(images))) +\
                     " === Test-Err:" + str(test_error/float(len(test_images)))[0:5] +\
                     " Test-Acc:" + str(test_correct_cnt/float(len(test_images))))
```
